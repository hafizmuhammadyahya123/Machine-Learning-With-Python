```python
import pandas as pd
df = pd.read_csv('/content/salary_data.csv')
df.head()
```

# **LOGISTIC REGRESSION IS CLASSIFICATION**

YAHAN OUTPUT CATEGORICAL MAIN HA MEANS (YES NO /TRUE FALSE /+ive -ive) YAHI REASON HA THIS IS CALEED BINARY CLASSIFICATION

```python
import numpy as np
import seaborn as sns

df = sns.load_dataset('tips')
df.head()

x = df[['total_bill','tip','size']]  #independent / input /data / feature
y = df['smoker']  #categorical / dependent / output / label

x.head()
y.head()

# import lib
from sklearn.linear_model import LogisticRegression
# call LogisticRegression
model = LogisticRegression()
# fit model / training model
model.fit(x,y)

# predict
print(model.predict([[40,1.01,5]]))
print(model.predict([[50,1.66,2]]))
```

# **NAIVE BAYES IS CLASSIFICATION**

YAHAN OUTPUT CATEGORICAL MAIN HA MEANS (YES NO /TRUE FALSE /+ive -ive) YAHI REASON HA THIS IS CALEED BINARY CLASSIFICATION

```python
import numpy as np
import seaborn as sns

df = sns.load_dataset('tips')
df.head()

x = df[['total_bill','tip','size']]  #independent / input /data / feature
y = df['smoker']  #categorical / dependent / output / label

x.head()
y.head()

# import naive bayes
from sklearn.naive_bayes import GaussianNB
# call GNB
model = GaussianNB()
# fit model / training model
model.fit(x,y)

# predict
print(model.predict([[40,1.01,5]]))
print(model.predict([[50,1.66,2]]))
```

```python
import numpy as np
import seaborn as sns

df = sns.load_dataset('tips')
df.head()

# encode categorical convert to numerical
df['smoker'] = df['smoker'].map({"Yes":1,"No":0})
df.head()

x = df[['total_bill','smoker','size']]  #independent / input /data / feature
y = df['tip']  #dependent / output / label

x.head()
y.head()

# import Decision Tree Regressor
from sklearn.tree import DecisionTreeRegressor
# call DecisionTreeRegressor
model = DecisionTreeRegressor()
# fit model / training model
model.fit(x,y)

# # predict
print(model.predict([[40,1,5]]))
print(model.predict([[50,0,2]]))
```

# **REGRESSION MATRICS**

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
df = sns.load_dataset('tips')
df
```

```python
x = df[['total_bill','size']]
y = df['tip']
```

```python
x.head()
```

```python
y.head()
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
model = LinearRegression()
model = model.fit(x_train,y_train)
model
```

```python
from sklearn.metrics import mean_squared_error,r2_score,mean_absolute_error
y_pred = model.predict(x_test)
```

```python
y_test
```

```python
print(f"MEAN SQUARED ERROR:{mean_squared_error(y_test,y_pred)}")
print(f"r2 score:{r2_score(y_test,y_pred)}")
print(f"MEAN ABSOLUTE ERROR:{mean_absolute_error(y_test,y_pred)}")
# root mean squared error
print(f"ROOT MEAN SQUARED ERROR:{np.sqrt(mean_squared_error(y_test,y_pred))}")
```
