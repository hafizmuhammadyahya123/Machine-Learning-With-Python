```python
# drop deck column
df.drop('deck',axis=1,inplace=True)
df.head()
```

```python
# impute missing values with mean
df['age'] = df['age'].fillna(df['age'].mean)

# check the num of missing values in each column
df.isnull().sum().sort_values(ascending = False)
```

```python
# impute missing values with median
df['embarked'] = df['embarked'].fillna(df['embarked'].median)

# check the num of missing values in each column
df.isnull().sum().sort_values(ascending = False)
```

```python
# impute missing values with mean
df['embark_town'] = df['embark_town'].fillna(df['embark_town'].mode()[0])

# check the num of missing values in each column
df.isnull().sum().sort_values(ascending = False)
```

# MULTIPLE IMPUTATION BY CHAINED EQUATION (MICE)

```python
# import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.experimental import enable_iterative_imputer
from sklearn.impute import IterativeImputer

# load data set
df = sns.load_dataset('titanic')
df.head()
```

```python
df.isnull().sum().sort_values(ascending = False)
```

```python
# encode the data using Label encoding
from sklearn.preprocessing import LabelEncoder

# first Method
encoder = LabelEncoder()
df['sex'] = encoder.fit_transform(df['sex'])
df['embarked'] = encoder.fit_transform(df['embarked'])
df['class'] = encoder.fit_transform(df['class'])
df['who'] = encoder.fit_transform(df['who'])
df['deck'] = encoder.fit_transform(df['deck'])
df['embark_town'] = encoder.fit_transform(df['embark_town'])
df['alive'] = encoder.fit_transform(df['alive'])

df.head()
```

```python
# second method
# encode the data using label encoding
from sklearn.preprocessing import LabelEncoder

# column to encode
column_to_encode = df[['sex','embarked','class','who','deck','embark_town','alive']]

# Dictionary to store Label Encoder each column for encoding
label_encoders = {}

# Loop to appply label encoder to each column for encoding
for col in column_to_encode:
  # Create a new label encoder for the column
  le = LabelEncoder()
  # fit & transform the data
  df[col] = le.fit_transform(df[col])
  # store the encoder from the dictionary
  label_encoders[col] = le

df.head()
```

```python
df.isnull().sum().sort_values(ascending=False)
```

```python
# IMPUTE THE MISSING VALUES WITH ITERATIVE IMPUTER
# CALL THE ITERATIVE IMPUTER CLASS WITH MAX ITERATOR
imputer = IterativeImputer(max_iter=10)

# IMPUTE MISSING VALUES WITH ITERATIVE IMPUTER IN THE for loop FOR age,embark_town,embarked columns

# column to impute
columns_to_impute = ['age','embark_town','embarked','deck']

# loop to impute each column
for col in columns_to_impute:
  df[col] = imputer.fit_transform(df[[col]])

# check the missing values
df.isnull().sum().sort_values(ascending=False)
```
