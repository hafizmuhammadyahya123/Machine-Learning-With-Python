# REGRESSION METRICS

```python
import seaborn as sns
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error,mean_absolute_error,mean_absolute_percentage_error,r2_score
```

```python
df = sns.load_dataset('tips')

df['smoker_num'] = df['smoker'].map({'Yes':1,'No':0})

x = df[['total_bill','smoker_num','size']] #indep / features
y = df['tip'] #dep / label

x_train,x_test,y_train,y_test = train_test_split(x,y,train_size=0.8,random_state=42)

model = LinearRegression().fit(x_train,y_train)

machine_prediction = model.predict(x_test)

print('mean_squared_error',mean_squared_error(y_test,machine_prediction))
print('mean_absolute_error',mean_absolute_error(y_test,machine_prediction))
print('mean_absolute_percentile_error',mean_absolute_percentage_error(y_test,machine_prediction))
print('root_mean_squared_error',np.sqrt(mean_squared_error(y_test,machine_prediction)))
print('r2_score',r2_score(y_test,machine_prediction))
```

# REGRESSION metrics

```python
import numpy as np
import seaborn as sns

df = sns.load_dataset('tips')

df['smoker_num'] = df['smoker'].map({'Yes':1,'No':0})

x = df[['total_bill','smoker_num','size']]
y = df['tip']

from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42)

from sklearn.linear_model import LinearRegression
model = LinearRegression().fit(x_train,y_train)

y_cap = model.predict(x_test)


from sklearn.metrics import mean_squared_error,mean_absolute_error,mean_absolute_percentage_error,r2_score
print(mean_squared_error(y_test,y_cap))
print(mean_absolute_error(y_test,y_cap))
print(np.sqrt(mean_squared_error(y_test,y_cap)))
print(mean_absolute_percentage_error(y_test,y_cap))
print(r2_score(y_test,y_cap))
```
